# Feature **Selection**

```python
# Correlation with target
corr = df.corr()["Diabetes_binary"].sort_values(ascending=False)

# Select features with correlation threshold
selected_corr = corr[abs(corr) > 0.05]

print("Selected based on correlation:\n")
print(selected_corr)

# Extract feature names
selected_features_corr = selected_corr.index.tolist()
selected_features_corr.remove("Diabetes_binary")
```

Remove **multicollinearity**

```python
# Correlation matrix
corr_matrix = df[selected_features_corr].corr().abs()

upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# Identify highly correlated features
to_drop = [col for col in upper.columns if any(upper[col] > 0.8)]

print("\nHighly correlated features to drop:\n", to_drop)

# Final features after removing redundancy
selected_features = [f for f in selected_features_corr if f not in to_drop]

print("\nFeatures after removing redundancy:\n")
print(selected_features)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Use only selected features
X = df[selected_features]
y = df["Diabetes_binary"]

# Apply Chi-Square
selector = SelectKBest(score_func=chi2, k=10)
X_new = selector.fit_transform(X, y)

# Final selected features
selected_features_final = X.columns[selector.get_support()]

print("\nFinal Selected Features:\n")
print(selected_features_final)
```

```python
#Chi-Square Scores
chi_scores = selector.scores_

chi_df = pd.DataFrame({
    "Feature": X.columns,
    "Chi2 Score": chi_scores
})

chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

print("\nChi-Square Feature Importance:\n")
print(chi_df)
```

```python
#visualization
plt.figure(figsize=(10,5))

plt.barh(chi_df["Feature"], chi_df["Chi2 Score"])

plt.title("Chi-Square Feature Importance")
plt.xlabel("Score")

plt.gca().invert_yaxis()

plt.show()
```
